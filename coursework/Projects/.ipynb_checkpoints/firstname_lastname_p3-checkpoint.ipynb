# Project 3: Poisonous mushrooms

In this project, you'll investigate properties of mushrooms. This classic dataset contains over 8000 observations, where each mushroom is described by a variety of features like color, odor, etc., and the target variable is an indicator for whether the mushroom is poisonous. Since all the observations are categorical, I've binarized the feature space. Look at the feature_names below to see all 126 binary names.

You'll start by running PCA to reduce the dimensionality from 126 down to 2 so that you can easily visualize the data. In general, PCA is very useful for visualization (though sklearn.manifold.tsne is known to produce better visualizations). Recall that PCA is a linear transformation. The 1st projected dimension is the linear combination of all 126 original features that captures as much of the variance in the data as possible. The 2nd projected dimension is the linear combination of all 126 original features that captures as much of the remaining variance as possible. The idea of dense low dimensional representations is crucial to machine learning!

Once you've projected the data to 2 dimensions, you'll experiment with clustering using KMeans and density estimation with Gaussian Mixture Models. Finally, you'll train a classifier by fitting a GMM for the positive class and a GMM for the negative class, and perform inference by comparing the probabilities output by each model.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
# deprecated: from sklearn.mixture import GMM
from sklearn.mixture import GaussianMixture as GMM

from matplotlib.colors import LogNorm
```

Load feature names.

```python
feature_names = []
with open('../Data/mushroom.map') as fmap:
    for line in fmap:
        [index, name, junk] = line.split()
        feature_names.append(name)

print('Loaded feature names:', len(feature_names))
#print feature_names
```

Load data. The data is sparse in the input file, but there aren't too many features, so we'll use a dense representation, which is supported by all sklearn objects.

```python
X, Y = [], []

with open('../Data/mushroom.data') as fdata:
    for line in fdata:
        items = line.split()
        Y.append(int(items.pop(0)))
        x = np.zeros(len(feature_names))
        for item in items:
            feature = int(item.split(':')[0])
            x[feature] = 1
        X.append(x)

# Convert these lists to numpy arrays.
#print len(X), len(Y)
X = np.array(X)
Y = np.array(Y)

# Split into train and test data.
train_data, train_labels = X[:7000], Y[:7000]
test_data, test_labels = X[7000:], Y[7000:]

# Check that the shapes look right.
print(train_data.shape, test_data.shape)
```

```python
def P1():
### STUDENT START ###
    n_comp = 50
    pca_mod = PCA(n_components = n_comp)
    pca_mod.fit(train_data)
    print(pca_mod.explained_variance_ratio_)
### STUDENT END ###

P1()
```

(3) Run KMeans with [1,16] clusters over the 2d projected data. Mark each centroid cluster and plot a circle that goes through the most distant point assigned to each cluster.

```python
def P3():
### STUDENT START ###
    np.random.seed(0)
    # Create 2d projected data
    pca_mod = PCA(n_components = 2)
    data2d = pca_mod.fit_transform(train_data)
    #data2d = pca_mod.components_.T
    
    # Create color map for 16 clusters
    cmap = plt.get_cmap('jet')
    colors = cmap(np.linspace(0, 1, 16))

    #####################################################
    def distance(p1, p2):
        return np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2 )

    
    def Get2DClusters (ax, Xpair, N):
        km = KMeans (n_clusters=N, init='k-means++')
        clstrs = km.fit(Xpair)

        myClCols = np.empty(Xpair.shape[0], 'object') #, 's256') #"string")
        for ii in np.arange (0, len(clstrs.labels_)):
            color_index = clstrs.labels_[ii]
            myClCols[ii] = colors[color_index]#myColors[color_index]

        ax.scatter(Xpair[:,0], Xpair[:,1], c = myClCols)

        myCntrCol = np.empty(N, 'object')
        for ii in np.arange (0, N):
            color_index = ii
            myCntrCol[ii] = colors[color_index]

        ax.scatter (clstrs.cluster_centers_[:,0], clstrs.cluster_centers_[:,1], marker='x', c = colors, s = 100)
        #print("clstrs.inertia_", clstrs.inertia_)
        ax.set_title (str(N) + " clusters")
        
        # Now, for each center, find the circle that goes through the farthest poin in the cluster
        for i in np.arange(0, N):
            max_distance = 0.0
            max_point = clstrs.cluster_centers_[i]
            for j in np.arange(0, len(clstrs.labels_)):
                # this belongs to cluster i
                if (clstrs.labels_[j] == i):
                    point = [Xpair[j,0], Xpair[j,1]]
                    d = distance(clstrs.cluster_centers_[i], point)

                    if (d > max_distance):
                        max_distance = d
                        max_point = point
                                               
            ax.add_artist(plt.Circle(clstrs.cluster_centers_[i], max_distance, color=colors[i], fill=False))       

        return clstrs
    #####################################################
    
    fig, axes = plt.subplots(nrows=4, ncols=4, sharex=False, sharey=False, 
                             figsize = (16,16))
    n = 0
    for i in range(4):
        for j in range(4):
            ax=axes[i,j]
            n += 1
            clstrs = Get2DClusters(ax, data2d, n)
    plt.show()

### STUDENT END ###

P3()
```

(4) Fit a Gaussian Mixture Model for the positive examples in your 2d projected data. Plot the estimated density contours as shown here: http://scikit-learn.org/stable/auto_examples/mixture/plot_gmm_pdf.html#example-mixture-plot-gmm-pdf-py. Vary the number of mixture components from 1-4 and the covariance matrix type ('spherical', 'diag', 'tied', 'full').

```python
def P4():
### STUDENT START ###
    
    def plot_gmm(ax, gmm_n_components, data2d, cov_type):
        gm_mod = GMM(n_components = gmm_n_components, covariance_type=cov_type)
        gm_mod.fit(data2d)

        # display predicted scores by the model as a contour plot
        x = np.linspace(-6., 6.)
        y = np.linspace(-6., 6.)
        #x = np.linspace(-20., 30.)
        #y = np.linspace(-20., 40.)
        X, Y = np.meshgrid(x, y)
        XX = np.array([X.ravel(), Y.ravel()]).T
        Z = -gm_mod.score_samples(XX)
        Z = Z.reshape(X.shape)

        CS = ax.contour(X, Y, Z, norm=LogNorm(vmin=1.0, vmax=1000.0),
                         levels=np.logspace(0, 3, 10))
        #plt.colorbar(CS, shrink=0.8, extend='both')
        fig.colorbar(CS, ax=ax)

        ax.scatter(data2d[:, 0], data2d[:, 1])

        ax.set_title('# comps = '+ str(gmm_n_components) + ', cov=' + cov_type)
    
    # this procedure project data onto a 2-d PCA model and then plot the estimated contours 
    def project_data_plot(axes, gmm_n_components, data):
        pca_mod = PCA(n_components = 2)
        pca_mod.fit(train_data)
        data2d = pca_mod.transform(data)

        #data2d = pca_mod.fit_transform(data)
        #data2d = pca_mod.components_.T
        i = 0
        for type in ['spherical', 'diag', 'tied', 'full']:
            ax=axes[gmm_n_components-1,i]
            plot_gmm(ax, gmm_n_components, data2d, type)
            i += 1

            
    
    # get positive data (poisonous)
    pos_data = train_data[train_labels == 1] # test_data[test_labels > 0] #
    
    fig, axes = plt.subplots(nrows=4, ncols=4, sharex=False, sharey=False, 
                             figsize = (16,16))
    for n in range(1,5):
        project_data_plot(axes, n, pos_data)

    plt.show()
### STUDENT END ###

P4()
```

(5) Fit two 4-component full covariance GMMs, one for the positive examples and one for the negative examples in your 2d projected data. Predict the test examples by choosing the label for which the model gives a larger probability (use GMM.score). What is the accuracy?

```python
def P5():
### STUDENT START ###

    def fit_gmm(data, label, n=4):
        # 4-component full covariance
        gm_mod = GMM(n_components = n, covariance_type='full')
        gm_mod.fit(data, label)
        return gm_mod

    # Split a 2d data into positive and negative sets
    def split_data(data, labels):
        data_pos = data[labels == 1]
        data_neg = data[labels == 0]
        return data_pos, data_neg
        
    def get_projected_data(data, n=2):
        pca_mod = PCA(n_components = n)
        pca_mod.fit(train_data)
        return pca_mod.transform(data)

    train_pos, train_neg = split_data(train_data, train_labels)
    test_pos, test_neg = split_data(test_data, test_labels)

    # Create 2d projected data
    train_pos_2d = get_projected_data(train_pos)
    train_neg_2d = get_projected_data(train_neg)
    test_pos_2d = get_projected_data(test_pos)
    test_neg_2d = get_projected_data(test_neg)

    gmm_positive = fit_gmm(train_pos_2d, train_labels[train_labels > 0])
    gmm_negative = fit_gmm(train_neg_2d, train_labels[train_labels == 0])

    print("Predicting positive test: score =", gmm_positive.score(test_pos_2d))
    print("Predicting negative test: score =", gmm_negative.score(test_neg_2d))
    print("Score for positive test is better than negative one, therefore it gives a larger probability")
    
    # Compute the weighted log probabilities for each sample.
    test_data_in_2D = get_projected_data(test_data)
    score_positive = gmm_positive.score_samples(test_data_in_2D)
    score_negative = gmm_negative.score_samples(test_data_in_2D)
    predicted_labels = (score_positive > score_negative).astype(int)
   
    print("Accuracy: %3.4f" % (float(sum(predicted_labels == test_labels))/test_labels.size))

### STUDENT END ###

P5()
```

(6) Ideally, we'd like a model that gives the best accuracy with the fewest parameters. Run a series of experiments to find the model that gives the best accuracy with no more than 50 parameters. For example, with 3 PCA components and 2-component diagonal covariance GMMs, you'd have:

( (3 mean vector + 3 covariance matrix) x 2 components ) x 2 classes = 24 parameters

You should vary the number of PCA components, the number of GMM components, and the covariance type.

```python
def P6():
### STUDENT START ###
    def run_model(pca_n_components, gmm_n_components, cov_type):
        
        def fit_gmm(data, label, n, cov_type):
            gm_mod = GMM(n_components = n, covariance_type=cov_type)
            gm_mod.fit(data, label)
            return gm_mod

        # Split a 2d data into positive and negative sets
        def split_data(data, labels):
            data_pos = data[labels > 0]
            data_neg = data[labels == 0]
            return data_pos, data_neg

        def get_projected_data(data, n):
            pca_mod = PCA(n_components = n)
            pca_mod.fit(train_data)
            return pca_mod.transform(data)

        def get_param_count(gmm_model):
            param_count = (np.prod(gmm_model.means_.shape) + 
                           np.prod(gmm_model.covariances_.shape) + 
                           np.prod(gmm_model.weights_.shape)) * 2
            return param_count
            

        ####################################################################
        train_pos, train_neg = split_data(train_data, train_labels)
        test_pos, test_neg = split_data(test_data, test_labels)

        # Create nd projected data
        pca_mod = PCA(n_components = pca_n_components)
        pca_mod.fit(train_data)


        train_pos_nd = pca_mod.transform(train_pos)
        train_neg_nd = pca_mod.transform(train_neg)
        test_pos_nd = pca_mod.transform(test_pos)
        test_neg_nd = pca_mod.transform(test_neg)

        gmm_positive = fit_gmm(train_pos_nd, train_labels[train_labels > 0], gmm_n_components, cov_type)
        gmm_negative = fit_gmm(train_neg_nd, train_labels[train_labels == 0], gmm_n_components, cov_type)

        parameter_count = get_param_count(gmm_positive)
        if (parameter_count > 50):
            print(pca_n_components, '\t\t', gmm_n_components, '\t%10s' % cov_type, '\t', parameter_count, '\t-> skipped')
            return parameter_count, 0.0
        

        # Compute the weighted log probabilities for each sample.
        test_data_in_nd = get_projected_data(test_data, pca_n_components)
        score_positive = gmm_positive.score_samples(test_data_in_nd)
        score_negative = gmm_negative.score_samples(test_data_in_nd)
        predicted_labels = (score_positive > score_negative).astype(int)

        accuracy = float(sum(predicted_labels == test_labels))/test_labels.size
#        print("For pca_n_components = %d, gmm_n_components = %d, covariance_type = %s: parameter count = %d, accuracy = %3.4f" 
#              % (pca_n_components, gmm_n_components, cov_type, parameter_count, accuracy))
        print(pca_n_components, '\t\t', gmm_n_components, '\t%10s' % cov_type, '\t', parameter_count, '\t%3.4f' % accuracy)
        return parameter_count, accuracy

    ##############################
    # trying different settings
    best_pca_n = 0
    best_gmm_n = 0
    best_parameter_count = 0
    best_accuracy = 0.0
    best_cov_type = ''
    print("PCA comps, GMM comps, Covariance, Parameter Count, Accuracy rate")
    print("----------------------------------------------------------------")
    for pca_n in range(2,10):
        for gmm_n in range(1,6):
            for cov_type in ('spherical', 'diag', 'tied', 'full'):
                parameter_count, accuracy = run_model(pca_n, gmm_n, cov_type)
                if parameter_count <= 50 and accuracy > best_accuracy:
                    best_pca_n = pca_n
                    best_gmm_n = gmm_n
                    best_parameter_count = parameter_count
                    best_accuracy = accuracy
                    best_cov_type = cov_type
    print("*********************************************************")            
    print("Model with best accuracy with no more than 50 parameters is with pca_n_components = %d, gmm_n_components = %d, covariance type = %s, accuracy = %3.4f" % 
        (best_pca_n, best_gmm_n, best_cov_type, best_accuracy))
            
### STUDENT END ###

P6()
```
